```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
iris= iris.rename(columns={'SepalLengthCm':'S_Length', 'SepalWidthCm':'S_Width',
                           'PetalLengthCm':'P_Length', 'PetalWidthCm':'P_Width'})
```

```python
sns.FacetGrid(iris, hue="Species",height=6).map(plt.scatter,"P_Length","S_Width").add_legend()
```

```python
sns.pairplot(iris[['S_Length','S_Width','P_Length','P_Width','Species']], hue="Species",diag_kind='kde')
```

```python
data=iris.values
X=data[:,0:4]
Y=data[:,5]
```

```python
print(X.shape)
print(Y.shape)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,Y, test_size=0.2)

print(X_train.shape)
print(X_train)

print(y_test.shape)
print(y_test)

print(X_test.shape)
print(X_test)

print(y_train.shape)
print(y_train)
```

```python
#SVM
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,y_train)

prediction1 = model_svc.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction1))
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeClassifier
model_DTC = DecisionTreeClassifier()
model_DTC.fit(X_train, y_train)

prediction3= model_svc.predict(X_test)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction3))
```

```python
flower_mapping = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris['Species']=iris['Species'].map(flower_mapping)
iris.head()
```

```python
X=iris [['S_Length','S_Width','P_Length','P_Width']].values
y= iris[['Species']].values

from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
model.fit(X,y)

model.score(X,y)
```

```python
expected = y
predicted = model.predict(X)
predicted

from sklearn import metrics

print(metrics.classification_report(expected, predicted))
```

```python
#CONFUSION MATRIX
print(metrics.confusion_matrix(expected, predicted))
```
